Chương trình đếm số lần xuất hiện của mỗi từ trong 1 file văn bản

```python
# Mở file
f = open("4300-0.txt", "r", encoding="utf-8")
```

```python
w_count = {}

# Doc tung dong cua file van ban
for line in f.readlines():
    # lay ra tung tu
    for w in line.split():
        if w in w_count:
            w_count[w] += 1
        else:
            w_count[w] = 1
print("Done.")
```
